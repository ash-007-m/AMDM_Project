```python
import matplotlib.pyplot as plt

# Multiply accuracy values by 100
df_train_metrics[['RELU_Train_Accuracy', 'EULU_Train_Accuracy', 'SELU_Train_Accuracy', 'GELU_Train_Accuracy', 'Softplus_Train_Accuracy']] *= 100
df_test_metrics[['RELU_Test_Accuracy', 'EULU_Test_Accuracy', 'SELU_Test_Accuracy', 'GELU_Test_Accuracy', 'Softplus_Test_Accuracy']] *= 100

# Create subplots for each metric
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(14, 18))

# Train Accuracy
df_train_metrics.plot(x='Epoch', y=['RELU_Train_Accuracy', 'EULU_Train_Accuracy', 'SELU_Train_Accuracy', 'GELU_Train_Accuracy', 'Softplus_Train_Accuracy'], ax=axes[0, 0])
axes[0, 0].set_title('Train Accuracy')
axes[0, 0].set_xlabel('Epoch')
axes[0, 0].set_ylabel('Accuracy (%)')

# Test Accuracy
df_test_metrics.plot(x='Epoch', y=['RELU_Test_Accuracy', 'EULU_Test_Accuracy', 'SELU_Test_Accuracy', 'GELU_Test_Accuracy', 'Softplus_Test_Accuracy'], ax=axes[0, 1])
axes[0, 1].set_title('Test Accuracy')
axes[0, 1].set_xlabel('Epoch')
axes[0, 1].set_ylabel('Accuracy (%)')

# Train Loss
df_train_metrics.plot(x='Epoch', y=['RELU_Train_Loss', 'EULU_Train_Loss', 'SELU_Train_Loss', 'GELU_Train_Loss', 'Softplus_Train_Loss'], ax=axes[1, 0])
axes[1, 0].set_title('Train Loss')
axes[1, 0].set_xlabel('Epoch')
axes[1, 0].set_ylabel('Loss')

# Test Loss
df_test_metrics.plot(x='Epoch', y=['RELU_Test_Loss', 'EULU_Test_Loss', 'SELU_Test_Loss', 'GELU_Test_Loss', 'Softplus_Test_Loss'], ax=axes[1, 1])
axes[1, 1].set_title('Test Loss')
axes[1, 1].set_xlabel('Epoch')
axes[1, 1].set_ylabel('Loss')

# Train Time
df_train_metrics.plot(x='Epoch', y=['RELU_Train_Time', 'EULU_Train_Time', 'SELU_Train_Time', 'GELU_Train_Time', 'Softplus_Train_Time'], ax=axes[2, 0])
axes[2, 0].set_title('Train Time')
axes[2, 0].set_xlabel('Epoch')
axes[2, 0].set_ylabel('Time (s)')

# Test Time
df_test_metrics.plot(x='Epoch', y=['RELU_Test_Time', 'EULU_Test_Time', 'SELU_Test_Time', 'GELU_Test_Time', 'Softplus_Test_Time'], ax=axes[2, 1])
axes[2, 1].set_title('Test Time')
axes[2, 1].set_xlabel('Epoch')
axes[2, 1].set_ylabel('Time (s)')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```
